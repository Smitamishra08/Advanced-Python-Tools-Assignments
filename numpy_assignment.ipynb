# Numpy Assignment

### This assignment demonstrates the use of Numpy for numerical computations.

#### The Task include:
- Creating arrays.
- Performing statistical calculations
- Manipulating matrices
- Performing arithmetic operations
- Reshaping arrays

```python
import numpy as np
```

```python
# 1. Create a 1D Array from 1 to 20:

arr = np.arange(1, 21)
print("Array:", arr)
```

```python
# A. Statistical Operations:

print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

```python
# B. Indices of elements greater than 10:

indices = np.where(arr > 10)
print("Indices where elements > 10:", indices)
```

```python
# Create 4x4 array from 1 to 16:

array2d = np.arange(1, 17).reshape(4,4)

print("\n4x4 Array:")
print(array2d)
```

```python
# Transpose:

print("\nTranspose:")
print(array2d.T)
```

```python
# row-wise sum:

print("\nRow-wise Sum:", np.sum(array2d, axis=1))
```

```python
# Column-wise Sum:

print("\nColumn-wise Sum:", np.sum(array2d, axis=0))
```

```python
# 3. Random 3x3 arrays

arr1 = np.random.randint(1, 21, (3,3))
arr2 = np.random.randint(1, 21, (3,3))

print("\nArray 1:")
print(arr1)

print("\nArray 2:")
print(arr2)

print("\nAddition:")
print(arr1 + arr2)

print("\nSubtraction:")
print(arr1 - arr2)

print("\nMultiplication:")
print(arr1 * arr2)

print("\nDot Product:")
print(np.dot(arr1, arr2))

```

```python
# Reshape & Slicing

arr3 = np.arange(1,13)

reshaped = arr3.reshape(3,4)

print("\nReshaped Array:")
print(reshaped)

print("\nSliced Array (first two rows and last two columns): ")
print(reshaped[:2, -2:])
```
