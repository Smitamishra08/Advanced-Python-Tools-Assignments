# Pandas Assignments

### This assignments demonstrates how to work with DataFrames in Pandas.

#### Task Include:
- Craeting DataFrames
- Data analysis
- Filtering and grouping data
- Sorting & exporting results

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)
```

```python
# A. First Five rows

print("\nFirst Five Rows")
print(df.head())
```

```python
# B. Summary Statistis
print("\nSummary Statistics")
print(df[['Age','Salary']].describe())
```

```python
# C. Average Salary of HR

hr_salary = df[df['Department']=='HR']['Salary'].mean()
print("\nAverage HR Salary:", hr_salary)
```

```python
# 2. Bonus Column

df['Bonus'] = df['Salary']* 0.10

print("\nDataFrame with Bonus:")
print(df)
```

```python
# 3. Filter age between 25 and 30

filtered = df[(df['Age']> 25) & (df['Age']<=30)]

print("\nEmployees age 25-30")
print(filtered)
```

```python
# 4. Group by Department

grouped = df.groupby('Department')['Salary'].mean()

print("\nAverage Salary by Department")
print(grouped)
```

```python
# 5.Sort by Salary

sorted_df = df.sort_values(by="Salary")

sorted_df.to_csv("sorted_salary.csv", index = False)

print("\nSorted DataFrame:")
print(sorted_df)
```
