# Matplotlib Assignments

### This assignment demonstarates different types of visualizations using Matplotlib.

#### Plots Created
- Line Plot
- Bar Graph
- Pie Chart
- Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Line Plot
x = [1,2,3,4,5]
y = [10,15,25,30,50]

plt.plot(x,y)
plt.title("Line Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()
```

```python
# Bar Graph

students = ['John', 'Jane','Alice','Bob']
marks = [75,85,60,90]

plt.bar(students, marks, color=['blue', 'green', 'orange', 'red'])
plt.title("Student Marks")
plt.show()
```

```python
# Pie Chart

regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45,25,20,10]

plt.pie(revenue, labels=regions, explode=[0.1,0,0,0], autopct='%1.1f%%')
plt.title("Revenue Distribution")
plt.show()
```

```python
# 4. Histogram

random_numbers = np.random.randint(1, 100, 1000)

plt.hist(random_numbers, bins=20)
plt.title("Histogram of Random Numbers")
plt.show()
```
